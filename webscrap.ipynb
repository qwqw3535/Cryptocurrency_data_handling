```python
import requests
from bs4 import BeautifulSoup
import json
from datetime import date, timedelta, datetime
import pandas as pd
import time
import os
```

```python
initial_date = date(2018,1,7)
date_list = [initial_date]
for i in range(272):
    new_date = date_list[-1] + timedelta(weeks=1)
    date_list.append(new_date)
for i in date_list:
    i = i.strftime("%Y%m%d")
print(date_list)
```

```python
for date in date_list:
    url = f"https://coinmarketcap.com/historical/{date.strftime('%Y%m%d')}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find_all('tbody')[0]

    data = []
    rows = table.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        print(cols)
        data.append({
            'symbol': cols[2],
            'market_cap': cols[3],
            'volume': cols[6],
            '1d_change': cols[8],
            '7d_change': cols[9],
        })
    with open(f"snapshot/{date.strftime('%Y%m%d')}_snapshot.json", 'w') as f:
        json.dump(data, f)
```

```python
def extract_week(file_path, binance_pairs, ranking_df):
    df = pd.read_json(file_path)
    df = df.dropna()
    df = df[df['symbol'].isin(binance_pairs)]
    df['percent_change_7d_rank'] = df['percent_change_7d'].rank(ascending=False)
    df['market_cap_rank'] = df['market_cap'].rank(ascending=True)
    df['ICO_date_rank'] = df['date_added'].rank(ascending=False)
    df['ranking_sum'] = df['percent_change_7d_rank'] + df['market_cap_rank']
    df = df.sort_values(by=['ranking_sum'])
    file_name = file_path.split('/')[-1].split('_')[0]
    df = df.reset_index(drop=True)
    for i in range(7):
        temp_date = file_name + timedelta(days=i)
        date_obj = datetime.datetime.strptime(temp_date, "%Y%m%d")
        timestamp_millisec = int(date_obj.timestamp() * 1000)
        ranking_df.loc[timestamp_millisec, :] = df.loc[:20,'symbol']
    return ranking_df
    
    
```

```python
binance_path = "C:/Users/qwqw3/Desktop/freqtrader/user_data/data/binance"
binance_pairs = []
for (root, dirs, files) in os.walk(binance_path):
    for file in files:
        symbol = file.split("_")[0]
        if (symbol not in binance_pairs) and ("DOWN" not in symbol) and ("UP" not in symbol):
            binance_pairs.append(symbol)
```
